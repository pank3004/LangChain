```python
import random

class NakliLLM: 
    def __init__(self): 
        print("LLM created")

    def predict(self, prompt): 
        response_list = [
        'Delhi is the capital of India',
        'IPL is a cricket league',
        'AI stands for Artificial Intelligence'
        ]

        return {'response': random.choice(response_list)}
```

```python
llm=NakliLLM()
```

```python
llm.predict('what is capital of india')
```

```python
class NakliPromptTemplate: 
    def __init__(self, template, input_variables): 
        self.template=template
        self.input_variables=input_variables

    def format(self, input_dict): 
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}',
    input_variables=['length', 'topic']
)
```

```python
prompt = template.format({'length':'short','topic':'india'})
```

```python
prompt
```

```python
llm=NakliLLM()
```

```python
llm.predict(prompt)
```

```python
class NakliLLMChain: 
    def __init__(self, llm, prompt): 
        self.llm=llm
        self.prompt=prompt
    def run(self, input_dict): 
        final_prompt=self.prompt.format(input_dict)
        result=self.llm.predict(final_prompt)
        return result['response']
    
```

```python
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}',
    input_variables=['length', 'topic']
)
```

```python
llm=NakliLLM()
```

```python
chain=NakliLLMChain(llm, template)
chain.run({'length':'short', 'topic': 'india'})
```

Earlier (in 2023), LangChain used to have separate systems for different parts — like:
(these chain are created as i created NakliLLMChain)

LLMChain

SequentialChain

SimpleSequentialChain

RouterChain

etc.

These were confusing, hard to combine, and inconsistent — because every class behaved a bit differently.

So, the LangChain tech team decided to unify everything under one simple interface called Runnable.

Now, whether you’re chaining models, prompts, retrievers, or even custom functions — all follow the same pattern:

output = my_runnable.invoke(input)
